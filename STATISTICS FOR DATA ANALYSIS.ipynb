**1. TYPES OF STATISTICS:**
    A. Descriptive Statistics
    B. Inferential Statistics
    
**A. Descriptive Statisctics**

*Measure of Central Tendency
    1. Mean
    2. Median
    3. Mode
    4. Skewness

*Measure of Variabiltiy
    1. Range
    2. Variance / Std Deviation
    3. Dispersion
    4. Quantiles (Values that divide the number of data points into 4 more or less equal parts, or quarters)
    
**B. Inferential Statistics**

    1.Population : Set of resources from where we can collect data
    
    2.Sample : Is nothing but a subset of Population which is used for sampling of data and in inferential statistics to predict the outcome.
    
    3.Variable : Can be a number, a characteristics, or a quantity that can be counted.It can be also called a data point.


**2. FIVE POINT SUMMARY**

1. The minimum
2. Q1 (the first quartile , or the 25% mark)
3. The median
4. Q3 (the third quartle, or the 75% mark)
5. The maximum
